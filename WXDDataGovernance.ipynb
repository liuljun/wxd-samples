```python
import prestodb
import pandas as pd

# PrestoDB connection parameters
presto_host = 'the hostname of Presto engine'
presto_port = 30658  # Replace with your PrestoDB port
presto_user = 'ibmlhapikey'
presto_catalog = 'iceberg_data'
presto_schema = 'sales'
presto_password = 'your IAM apikey'

# Additional SSL options
ssl_options = {'SSL': 'true'}

# Query to execute
query = 'SELECT C_CUSTOMER_SK, C_FIRST_NAME,C_LAST_NAME  FROM "iceberg_data"."sales"."customer" LIMIT 10'

# Connect to PrestoDB

conn =  prestodb.dbapi.connect(
      host=presto_host,
      port=presto_port,
      catalog=presto_catalog,
      schema=presto_schema,
      user=presto_user,
      http_scheme='https',
      auth=prestodb.auth.BasicAuthentication(presto_user,presto_password),
) 


# Create a PrestoDB cursor
cur = conn.cursor()

# Execute the query
cur.execute(query)

# Fetch and print the results
results = cur.fetchall()
df = pd.DataFrame(results, columns=[desc[0] for desc in cur.description])
print(df)

# Close the cursor and connection
cur.close()
conn.close()
```
